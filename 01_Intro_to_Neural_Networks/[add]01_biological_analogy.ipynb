## Additional material to the Lesson #1. Analogy between artificial and biological neural network

##### Author: [Tatiana Gaintseva](https://www.linkedin.com/in/tgaintseva/)

Here we will try to show how to draw an analogy between an artificial and biological neurons and a neural network.


First, **we will describe in general terms how a biological neural network and individual neurons in our brain are structured**.

**Disclaimer:** our description of biological processes does not claim to be complete and rigorous: we will describe them at a very abstract, conceptual level.

So, our brain consists of billions of neurons connected to each other in a network. The connections between neurons are called **synapses**: information is transmitted through them. Neurons are connected to each other quite chaotically, there is no special structure in the connections. Moreover, new connections can appear between neurons and old ones can disappear.

Each neuron receives information from one group of neurons and transmits it to another group of neurons. Information in the brain is presented in the form of electrical impulses, so when a neuron receives and transmits information, an electric current passes through it and it "lights up".

Also, the synapses between neurons have different resistance. The resistance of the connection determines how much current will pass from one neuron to another. If the resistance is high, only a small part of the impulse released by the first neuron will reach the second. If the resistance is low, then almost the entire impulse will pass through the connection without hindrance. Thus, **the resistance value can be considered the importance of the signal**: how important the signal from the first neuron is for the second neuron.

This is how the process of information transfer between neurons in the brain is generally structured.

It is clearly seen how individual layers of neurons light up and go out in sequence: this is how information flows through the neural network of the brain from one part to another.

Now let's discuss the structure of an individual biological neuron in a little more detail. In general, **a neuron has three components: dendrites, an axon, and a membrane**.

Neuron.jpg

**Dendrites** are the inputs of a neuron, through which the neuron receives information from other neurons in the brain.

The **membrane and cytoplasm** process the information received through the dendrites: all impulses from other neurons are added up, and the sum is compared with a certain threshold value. If the sum is greater than the threshold, the neuron is activated and the information is transmitted further to other neurons. If the sum is less than the threshold, then the information received is considered insignificant, the neuron is not activated and does not transmit information further along the network. That is, the impulse goes out.

And the last element of the neuron is the **axon**. The axon connects to the dendrites of other neurons by synapses. It is through it that the information accumulated in the neuron is transmitted to the next neurons. Well, if activation has occurred, of course.

This is approximately how a biological neuron is structured. Now **let's build a model of an artificial neuron based on its structure**. It'll look like that:

The green circle represents membrane, the edges on the left are the synapses through which information from other neurons marked with gray circles enters the nucleus. The edges on the right are the synapses through which information goes to other neurons further along the network.

In a computer, all information is represented as numbers. Therefore, our artificial neuron will receive and transmit information in the form of numbers, and not electrical impulses, as happens in a biological neural network.

This is how an artificial neuron processes information: lour neuron is connected to five other neurons: it receives information from three and passes it on to two. Let the three incoming neurons send pulses of sizes $X^1, X^2$ and $X^3$ as the input of our green neuron. Each edge between neurons has a weight: some real number $w^i$. When a signal from a neuron passes along an edge, the signal is multiplied by the edge weight. That is, three signals will come to our green neuron: $X^1w^1, X^2w^2, X^3w^3$. The **edge weights in an artificial neuron are analogous to the resistance on the connections between neurons** in a biological neural network.
Then, inside the core, these signals are summed up together. And then this signal is compared with the threshold value $thr$. If the accumulated signal is greater than $thr$, the neuron "fires" and the signal is passed further.

**We see that such a model of artificial neuron resembles logistic regression and the artificial neuron we talked about in class**. The only difference is the activation function. In an artificial neural network, it is not threshold-like.

This is how you can draw an analogy between a biological and artificial neuron and a neural network. This analogy, however, does not explain why in an artificial neural network neurons are combined into layers, and connections exist only between neurons of successive layers. But we have already figured out why the artificial network is structured this way in the main part of the lesson.
